```python
# Process and visualize each stock
for stock_name, df in stock_data.items():
    print(f"Processing {stock_name}...")
    
    # Calculate indicators
    df = calculate_indicators(df)
    
    # Visualize data
    plot_with_sma(df, stock_name)
    plot_macd(df, stock_name)
```
